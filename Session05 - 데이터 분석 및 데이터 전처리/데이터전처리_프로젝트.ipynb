### 데이터프레임 컬럼 조회하기

```python
import pandas as pd
customerData = pd.read_csv("../dataset/customerData.csv")
```

```python
import numpy as np
```

```python
customerData["PRODUCTAGE_NEW"] = np.where(customerData.PRODUCTAGE < 1, 1, 0)
```

```python
customerData
```

```python
### 데이터 조작 방법
# 1. np.where을 활용하여 조작한다. (복잡한 로지에는 힘들어진다.)
# 2. 함수를 활용하여 조작한다. (복잡한 로직에도 강하다.)
# 3. loc 함수를 활용하여 조작한다. (가독성이 뛰어나지만 속도가 떨어진다.)
# 4. map함수를 활용하여 조작한다. (머신러닝에서 많이 사용된다.)
```

```python
##### 1. np.where 활용
```

```python
customerData["PRODUCTAGE_NEW"] = np.where(customerData.PRODUCTAGE < 1, 1,
                                         np.where(customerData.PRODUCTAGE < 2, 2, 3)) 
```

```python
customerData.loc[(customerData.PRODUCTAGE_NEW < 1) &
                 (customerData.PRODUCTAGE_NEW != 1)]
```

```python
##### 2. 함수 활용
```

```python
# 함수 정의 
# inputValue < 1 :1
# 1 <= inputValue < 2 : 2
# inputValue >= 2 : 3
```

```python
def valueClustering(inValue) : 
    
    outValue = inValue

    if inValue < 1 : 
        outValue = 1
    elif inValue < 2 :
        outValue = 2
    else :
        outValue = 3
    return outValue
```

```python
customerData["PRODUCTAGE_NEW"]=customerData.PRODUCTAGE.apply(valueClustering)
```

```python
##### 3. loc 방식 활용
```

```python
customerData.loc[customerData.PRODUCTAGE < 1, "PRODUCTAGE_NEW"] = 1
```

```python
customerData.loc[(customerData.PRODUCTAGE >= 1) &
                 (customerData.PRODUCTAGE < 2),
                "PRODUCTAGE_NEW"] = 2
```

```python
customerData.loc[(customerData.PRODUCTAGE >= 2), "PRODUCTAGE_NEW"] = 3
```

```python
customerData.loc[(customerData.PRODUCTAGE < 1) & 
                 (customerData.PRODUCTAGE_NEW != 1)]
```

```python
##### 4. map 함수 활용
```

```python
featuresData = pd.read_csv("../dataset/feature_regression_example.csv")
```

```python
ynMap = {"Y" : 1, "N" : 0}
```

```python
featuresData["HOLIDAY_NEW"] = featuresData.HOLIDAY.map(ynMap)
```

```python
customerData["NEW_COLUMN"] = customerData.CUSTID.str.lower() + "_" + customerData.EMI.astype(str)
```

### 집계함수

```python
# 무이자할부(EMI) 기준 평균지출비용(AVGPRICE)의 평균 산출
groupKey = "EMI"
meanValue= sortedData.groupby(by=groupKey)["AVGPRICE"].agg(["mean"])
```
