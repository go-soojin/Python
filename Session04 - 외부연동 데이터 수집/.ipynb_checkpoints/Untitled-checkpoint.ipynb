```python
rowList = []
columnList = []
columnNameList = []
```

```python
rows = xmlobj.find_all("item")
```

```python
xmlobj.find_all("item")
```

```python
# 각 행에 대한 반복문 수행
for j in range(0, len(rows)):

    eachRow = rows[j]
    columns = eachRow.find_all() # 한개의 행에 있는 모든 태그를 찾아라!

# 첫번째 내부 반복문 : 컬럼 값을 전부 담아라!!
    for i in range(0, len(columns)):
        eachColumn = columns[i]
       # 
        if j == 0:
            columnNameList.append(eachColumn.name)
        
        eachColumnValue = eachColumn.text
        columnList.append(eachColumn.text.replace(" ",""))

    rowList.append(columnList)
    columnList = []
```

```python
indata = pd.DataFrame(rowList)
```
