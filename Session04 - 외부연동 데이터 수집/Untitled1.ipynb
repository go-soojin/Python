```python
# 셀레니움을 활용해서 sparkkorea.com quiz에 접근
# 이후 spark 관련 퀴즈 타이틀과 링크정보를 수집하여 데이터프레임으로 변환
```

```python
import pandas as pd
from selenium import webdriver 

# 웹 브라우저 설정 및 브라우저 팝업
options = webdriver.ChromeOptions()

# 헤드리스 옵션 / 일반 옵션
options.add_argument("window-size = 1920*1080")
driverLoc = "../addon/chromedriver_win32/chromedriver.exe"
driver = webdriver.Chrome(executable_path=driverLoc, options = options)

# 웹페이지 파싱될 때까지 최대 3초 기다림
driver.implicitly_wait(3)
```

```python
targetUrl = "https://www.sparkkorea.com"
driver.get(targetUrl)
quizBtnXpath = '//*[@id="menu-item-382"]/a' 
quizBtn = driver.find_element_by_xpath(quizBtnXpath)
quizBtn.click()
finalUrl = driver.current_url
pgSource = driver.page_source
pgSource
```

```python
import bs4
bsObj = bs4.BeautifulSoup(pgSource, "html.parser")
QuizTag=bsObj.find(name ="div", attrs = {"id":"id_spark_quiz"} )
QuizTag
atags = QuizTag.findAll(name= "a")
atags
```

```python
TitleList = []
LinkList = []
```

```python
# Title 정보 수집
for i in range(0,len(atags)):
    Title = atags[i].text
    TitleList.append(Title)

# Link 정보  수집
for j in range(0,len(atags)):
    Link = atags[0].attrs["href"]
    LinkList.append(Link)
```

```python
df=pd.DataFrame(zip(TitleList,LinkList))
columnTitle = ["Title", "Link"]
df.columns = columnTitle
df
```

```python
df.to_csv("./quiz_result.csv", encoding="ms949", index=False)
```
